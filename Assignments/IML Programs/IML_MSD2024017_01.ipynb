**Code for Quick Sort**

In quicksort we take two pointers one from starting and and one from last and consider a PIVOT generally 1st element and traversing the array until the first pointer cross the value greator than pivot or second pointer reach a number smaller than pivot and then replace the pivot with second pointer value.

```python
def quicksort(arr):

    if len(arr) <= 1:
        return arr

    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quicksort(left) + middle + quicksort(right)

print(quicksort([3,6,8,10,1,2,1]))
```

**Slicing**

slicing is a process of splitting the data

```python
nums = list(range(5))    # range is a built-in function that creates a list of integers
print(nums)         # Prints "[0, 1, 2, 3, 4]"
print(nums[2:4])    # Get a slice from index 2 to 4 (exclusive); prints "[2, 3]"
print(nums[2:])     # Get a slice from index 2 to the end; prints "[2, 3, 4]"
print(nums[:2])     # Get a slice from the start to index 2 (exclusive); prints "[0, 1]"
print(nums[:])      # Get a slice of the whole list; prints ["0, 1, 2, 3, 4]"
print(nums[:-1])    # Slice indices can be negative; prints ["0, 1, 2, 3]"
nums[2:4] = [8, 9] # Assign a new sublist to a slice
print(nums)         # Prints "[0, 1, 8, 9, 4]"
```

**Code of Rank of Matrix Without Numpy**

```python
# without numpy

def swap(matrix, row1, row2):
    matrix[row1], matrix[row2] = matrix[row2], matrix[row1]

def divide(matrix, row, divisor):
    matrix[row] = [element / divisor for element in matrix[row]]

def subtract(matrix, target_row, source_row, multiplier):
    matrix[target_row] = [target - multiplier * source for target, source in zip(matrix[target_row], matrix[source_row])]

def findRank(matrix):
    num_rows = len(matrix)
    num_cols = len(matrix[0])
    row = 0

    for col in range(num_cols):
        if row >= num_rows:
            break
        pivot_row = row
        while pivot_row < num_rows and matrix[pivot_row][col] == 0:
            pivot_row += 1
        if pivot_row == num_rows:
            continue
        if pivot_row != row:
            swap(matrix, row, pivot_row)
        pivot_value = matrix[row][col]
        divide(matrix, row, pivot_value)
        for r in range(num_rows):
            if r != row:
                subtract(matrix, r, row, matrix[r][col])
        row += 1

    return sum(any(matrix[r]) for r in range(num_rows))



Matrix = [[2,4,1],
          [6,9,2],
          [3,5,1]]

print("Rank of the matrix:", findRank(Matrix))
```

**Code for finding the Order of matrix**

```python
# Order of a matrix

def matrix_order(matrix):
    if not matrix or not matrix[0]:
        return (0, 0)
    return (len(matrix), len(matrix[0]))



matrix = [
    [2,4,1],
    [6,9,2],
    [3,5,1]
]

order = matrix_order(matrix)
print("Order of the matrix:", order)
```

**Sort the data by pet**

There are two ways to sort.

By index
By value
By value means sorting according to value in a column.

In this example we sort the rows of the dataframe based on values in 'pet' column.

The parameter ascending = True means that we want the rows sorted in ascending order. This is the same as sql 'ASC'. To get descending order use ascending = False.

inplace is very important and you should always remember it. When inplace=True the dataframe is modified in place which means that no copies are made and your previous data stored in the dataframe is lost. By default inplace is always False. When it is false a copy is made of your data and that copy is sorted and returned as output.

The output of sort_values is always a dataframe returned but the behaviour depends strongly on the inplace parameter.

```python
df.iloc[0]
df.sort_values('pet',inplace=True, ascending=True)

df

df.loc[0] #index based
df.iloc[0] #relative position based indexing

df.iloc[:,2]

df.iloc[-1,:]
```

**Specifying Colors, Markers**

For the basic built-in colors, you can use a single letter

b: blue
g: green
r: red
c: cyan
m: magenta
y: yellow
k: black
w: white


```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4], [1,4,9,16])
plt.plot([0,1,5,7], [1,4,9,16], 'D-')
plt.plot([0,1,3,5], [3,4.5,7,18], 'ro-')
plt.show()

# Example of all markers
# From: https://stackoverflow.com/a/46637880
import matplotlib.pylab as plt
markers=['.',',','o','v','^','<','>','1','2','3','4','8','s','p','P','*','h','H','+','x','X','D','d','|','_']
descriptions=['point', 'pixel', 'circle', 'triangle_down', 'triangle_up','triangle_left', 'triangle_right', 'tri_down', 'tri_up', 'tri_left','tri_right', 'octagon', 'square', 'pentagon', 'plus (filled)','star', 'hexagon1', 'hexagon2', 'plus', 'x', 'x (filled)','diamond', 'thin_diamond', 'vline', 'hline']
x=[]
y=[]
for i in range(5):
    for j in range(5):
        x.append(i)
        y.append(j)
plt.figure(figsize=(12,6))
for i,j,m,l in zip(x,y,markers,descriptions):
    plt.scatter(i,j,marker=m)
    plt.text(i-0.15,j+0.15,s=m+' : '+l)
plt.axis([-0.1,4.8,-0.1,4.5])
plt.tight_layout()
plt.axis('off')
plt.show()
```
